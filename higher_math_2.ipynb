3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_dec(r, angle):
    angle = angle * np.pi/180
    return np.array([r*np.cos(angle), r*np.sin(angle)])
```

```python
points = np.linspace(0, 360, 200)
circle = polar_to_dec(1, points);
```

```python
from pylab import rcParams
plt.plot(circle[0], circle[1])
rcParams['figure.figsize']=5, 5;
```

Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def polar_circle(b, a_0 = 0, r_0 = 0):
    a = np.linspace(0, 360, 200)
    a = a * np.pi/180
    a_0 = a_0 * np.pi/180
    r = np.sqrt(b**2 + r_0**2 - 2*b*r_0*np.cos(a-a_0))
    return np.array([a, r])
```

```python
plt.polar(polar_circle(1)[0], polar_circle(1)[1])
plt.polar(polar_circle(2,a_0=45, r_0=0)[0], polar_circle(2,a_0=45, r_0=0)[1])
rcParams['figure.figsize']=5, 5
```
