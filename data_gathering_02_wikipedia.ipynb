В приложении парсинга википедии получить первую ссылку из раздела "Ссылки" 

и вывести все значимые слова из неё. Результат записать в файл в форматированном виде.

```python
from pprint import pprint
import requests
import re
```

```python

def get_topic():
    topic = input('О чем исследуем статью в Вики? ')
    return topic

def get_link():
    link='https://ru.wikipedia.org/wiki/'+get_topic().capitalize()
    return link

def get_topic_page():
    answer = ''
    article_text = ''
    while answer != 'N':
        link = get_link()
        response = requests.get(link)
        article_text = response.text
        if str(response) == '<Response [200]>':
            return article_text
        else:
            answer = input('Такой статьи, кажется, пока нет. Продолжим? Y/N')
            if answer != 'N':
                continue
            elif answer == 'N':
                print('До встречи!')
                break

```

```python
html = get_topic_page()
            
```

```python
def get_reference(html_content):
    words = re.findall(r'id="Ссылки">Ссылки.+\s.+\s.+\s.+\s.+\s.+\s.+\s.+\s.+<a rel="nofollow" class="external text" href="(.+)"', html_content)
#    words = re.findall(r'id="Ссылки">Ссылки[.\s]{0,}.+<a rel="nofollow" class="external text" href="(.+)"', html_content)
    return words
```

```python
reference_link = str(get_reference(html)[0])
```

```python
print(reference_link)
```

```python
reference_text = requests.get(reference_link).text
```

```python
def get_words(html_content):
    words = re.findall("[а-яА-Я]{3,}",html_content)
    return words

def get_common_words(topic):
    words_list = get_words(topic)
    rate={}
    for word in words_list:
        if word in rate:
            rate[word]+=1
        else:
            rate[word]=1
    rate_list = list(rate.items())
    rate_list.sort(key = lambda x: -x[1])
    return rate_list
```

```python

dict1 = get_common_words(reference_text)
pprint(dict1)
```
