Доработать приложение по поиску авиабилетов, чтобы оно возвращало билеты

по названию города, а не по IATA коду. (У aviasales есть для этого дополнительное API)

Пункт отправления и пункт назначения должны передаваться в качестве параметров. 

```python
from pprint import pprint
import requests
import json
```

```python
# функции запроса названий пунктов отправления и назначения и получение соответствующих IATA кодов
def get_city(m):
    if m == 0:
        a = input('Введите название города отправления')
    elif m == 1:
        a = input('введите название города прибытия')
    return a

def city_code():
    address = 'https://www.travelpayouts.com/widgets_suggest_params?q='
    link_city = f'{address}Из%20{get_city(0)}%20в%20{get_city(1)}'
    req_city = requests.get(link_city)
    json_answer = json.loads(req_city.text)
    return json_answer
```

```python
# вызов функции, запись соответствующего ответа в переменную
codes = city_code()
```

```python
print(codes)
```

```python
# поиск дешевых билетов с использованием полученных iata кодов во фразе запроса
service = 'http://min-prices.aviasales.ru/calendar_preload?'
fromCity = codes['origin']['iata']
toCity = codes['destination']['iata']
req_date=input('Дата вылета в формате YYYY-MM-DD: ')
link_ticket = f'{service}origin={fromCity}&destination={toCity}&depart_date={req_date}&one_way=true'
# запрос, запись ответа в переменную и сохранение в виде словаря
req_ticket = requests.get(link_ticket)
data = json.loads(req_ticket.text)
```

```python
# функция возврата названия города по IATA code
def city_name(m, dep=codes['origin']['name'], dest=codes['destination']['name']):
    address = 'http://autocomplete.travelpayouts.com/places2?'
    link = ''
    if m == 0:
        link = f'{address}term={dep}&locale=ru&types[]=country&callback=function' 
    elif m == 1:
        link = f'{address}term={dest}&locale=ru&types[]=country&callback=function'
    print(link)
    city_name = requests.get(link)
    json_answer = json.loads(city_name.text)
    return json_answer
```

```python
#Сделать форматированный вывод, который содержит в себе пункт отправления, пункт назначения,
# дату вылета, цену билета (можно добавить еще другие параметры по желанию)
n1 = 1
dep_city = codes['origin']['name']
dest_city = codes['destination']['name']
print(f'Пункт отправления {dep_city}, пункт назначения {dest_city}. Лучшая цена на запрашиваемую дату {req_date}:')
for i in data['best_prices']:
    if i['depart_date'] == req_date:
        value = i['value']
        changes = i['number_of_changes']
        print(f'{n1}. Цена билета {value}, количество пересадок {changes}')
        n1 += 1
```

```python
n2 = 1
pprint(f'Лучшие цены на другие даты:')
for i in data['best_prices']:
    date = i['depart_date']
    value = i['value']
    changes = i['number_of_changes']
    pprint(f'{n2}. Вылет {date},  цена билета {value}, количество пересадок {changes}')
    n2 += 1
```
